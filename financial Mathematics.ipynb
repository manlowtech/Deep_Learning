```python
"""
6. Interest Rates
Data Series: Federal Reserve Interest Rates, 1954-Present
Source: https://www.kaggle.com/datasets/federalreserve/interest-rates
About Dataset
Context
The Federal Reserve sets interest rates to promote conditions that achieve the mandate set by the Congress
— high employment, low and stable inflation, sustainable economic growth, and moderate long-term interest rates.
Interest rates set by the Fed directly influence the cost of borrowing money.
Lower interest rates encourage more people to obtain a mortgage for a new home or to borrow money for an automobile or for home improvement.
Lower rates encourage businesses to borrow funds to invest in expansion such as purchasing new equipment, updating plants, or hiring more workers.
Higher interest rates restrain such borrowing by consumers and businesses.
"""
data = pd.read_csv("index.csv")
data
```

```python
"""
performing exploratory data analysis (EDA) on the Federal Reserve Interest Rates dataset using Python.
"""
# Check for missing values and data types
print(data.info())
print(data.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Convert 'Date' to datetime
data['Date'] = pd.to_datetime(data['Year'])

# Plotting the Federal Funds Rate over time
plt.figure(figsize=(12, 6))
plt.plot(data['Date'], data['Effective Federal Funds Rate'], label='Effective Federal Funds Rate', color='blue')
plt.title('Federal Funds Rate Over Time (1954-Present)')
plt.xlabel('Year')
plt.ylabel('Interest Rate (%)')
plt.legend()
plt.grid()
plt.show()

"""
The Federal Funds Rate shows notable fluctuations over the decades,
reflecting monetary policy responses to economic conditions.
"""
```

```python
# Plotting the distribution of the Federal Funds Rate
plt.figure(figsize=(8, 6))
sns.histplot(data['Effective Federal Funds Rate'], bins=30, kde=True, color='orange')
plt.title('Distribution of Effective Federal Funds Rate')
plt.xlabel('Interest Rate (%)')
plt.ylabel('Frequency')
plt.grid()
plt.show()

"""
The distribution of the Federal Funds Rate indicates that
 most rates are concentrated around the lower end, with fewer occurrences of very high rates.
"""
```

```python
# Creating a 'Decade' column
data['Decade'] = (data['Date'].dt.year // 10) * 10

# Boxplot of Federal Funds Rate by Decade
plt.figure(figsize=(12, 6))
sns.boxplot(x='Decade', y='Effective Federal Funds Rate', data=data, palette='Set2')
plt.title('Federal Funds Rate by Decade')
plt.xlabel('Decade')
plt.ylabel('Interest Rate (%)')
plt.xticks(rotation=45)
plt.grid()
plt.show()

"""
The boxplot illustrates how the median Federal
 Funds Rate has shifted over different decades, with significant spikes in the 1970s.
"""
```

```python
# Correlation matrix
correlation_matrix = data.corr()

# Plotting the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Heatmap of Interest Rates')
plt.show()
"""
Insight: The heatmap indicates that there is a strong correlation between different interest rate types,
suggesting that changes in one rate are often mirrored by changes in others.
"""
```

```python
# Scatter plot of Federal Funds Rate vs. Inflation Rate
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Inflation Rate', y='Effective Federal Funds Rate', data=data)
plt.title('Federal Funds Rate vs. Inflation Rate')
plt.xlabel('Inflation Rate (%)')
plt.ylabel('Federal Funds Rate (%)')
plt.grid()
plt.show()

"""
The scatter plot reveal a relationship between inflation and the Federal Funds Rate,
suggesting that higher inflation could lead to an increase in interest rates.
"""
```

```python
"""
TASK 6


About Dataset
Title:
VIX (^VIX) Historical Data from Yahoo Finance
Daily historical data for the VIX (^VIX) from 2000 to TODAY
Data Source: https://www.kaggle.com/datasets/guillemservera/vix-cboe-volatility-index-daily-updated
Description:
This dataset contains the historical daily prices of the Volatility Index, VIX (^VIX), sourced from Yahoo Finance using the yfinance library. The data spans from the start of the year 2000 until TODAY.

Introduction:
The VIX Index, colloquially known as the "fear index", measures the market's expectation of volatility over the next 30 days. This dataset provides daily data on the VIX, offering insights into market perceptions of future volatility. It's especially valuable for financial analysts, traders, and researchers seeking to understand market dynamics, evaluate potential risks, or craft trading strategies.

File Contents:

vix_data.csv: Contains the daily open, high, low, close prices, and volume for the VIX.
"""

vix_data = pd.read_csv("vix_daily.csv")
vix_data

illiquid_security_data = pd.read_csv("AAPL.csv")
illiquid_security_data
illiquid_security_data.rename(columns = {'Date':'date'}, inplace = True)

# Merge the datasets on the date
merged_data = pd.merge(vix_data, illiquid_security_data, on='date')

# Scatter plot of VIX vs. Illiquid Security Volume
plt.figure(figsize=(10, 6))
sns.scatterplot(x='close', y='Volume', data=merged_data)
plt.title('VIX vs. Illiquid Security Trading Volume')
plt.xlabel('VIX Index')
plt.ylabel('Trading Volume')
plt.grid()
plt.show()
"""
The scatter plot indicates that as the VIX increases, the trading volume of the illiquid security tends to decrease,
demonstrating that higher market volatility leads to greater illiquidity.
"""
```
